Packaging History:
* DL tarbal, run setup.py
* Pypi was only an index, not a host
* Packages were hosted everywhere
* Single server in Sweden...
* Globally installed package
* People just copied things into site-packages manually
* Poor user experience
* easy_install was next iteration -- grabbed from CheeseShop, installed into system
  * better overall, but no easy_uninstall
* 2010 onward -- pip replaces easy_install, virtualenv common practice
* requirements.txt files become standard
* virtualenv: creates isolated python homes, one for each project -- unique to Python community
* pip: resolves, downloads, installs and uninstalls packages from arbitrary urls
  * still used today
* node.js, php, rust and ruby all have lockfiles, but not python
* virtualenv problems:
  * difficult to understand, barrier to entry, manual process (virtualenv-wrapper makes this easier to use)
* way to snag curent eng: $ pip freeze > requirements.txt
* requirements.txt -- do you do just "Flask" or "Flask + all dependencies called out"?


Change req.txt: 2 files (see pip-tools):
* requirements-to-freeze.txt (Flask)
* requirements.txt (everything)
* this is NOT a real solution

**Pipfile will replace requirements.txt**
* TOML, easy to read/write manually
* two groups: packages and dev-packages
* eventually landing in pip proper
* Pipfile.lock, generated by Pipfile:
  * JSON
  * Includes hashes of wheels, sdists, etc.

pipenv -- since Pipfile not yet integrated into pip, use pipenv!
* lets you use Pipfile + Pipfile.lock today
* automates away virtualenv
* it's the officially recommended tool
